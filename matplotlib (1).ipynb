```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#1

plt.suptitle("relationship b/w x & y",color="brown",fontsize="20")
x=np.array([1,2,3])
y=np.array([2,4,6])
plt.subplot(2,2,1)
plt.plot(x,y,"o",ms=5,mec="r",mfc="hotpink")
plt.grid(color="g",ls="dotted",lw=1)

#2
x=np.array([5,10,15])
y=np.array([30,20,10])
plt.subplot(2,2,2)
plt.plot(x,y,"x",ms=5,mec="r",mfc="hotpink")
plt.grid(color="g",ls="dotted",lw=1)

#3
x=np.array([1,2,3])
y=np.array([2,4,6])
plt.subplot(2,2,3)
plt.plot(x,y,marker="*",ms=5,color="k")
#4
x=np.array([5,10,15])
y=np.array([30,20,10])
plt.subplot(2,2,4)
plt.plot(x,y,marker="*",ms=5,color="black")
plt.plot(x,y)
plt.show()

```
